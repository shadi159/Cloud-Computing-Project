````python
# @title Initialize AI Disease Model
import torch
from transformers import AutoImageProcessor, AutoModelForImageClassification
from PIL import Image
import io
import warnings

warnings.filterwarnings('ignore')

print("‚è≥ Loading Disease Detection Model...")

# We use the model you specified
MODEL_NAME = "linkanjarad/mobilenet_v2_1.0_224-plant-disease-identification"

try:
    processor = AutoImageProcessor.from_pretrained(MODEL_NAME)
    model = AutoModelForImageClassification.from_pretrained(MODEL_NAME)
    print("‚úÖ AI Model Loaded Successfully!")
except Exception as e:
    print(f"‚ùå Error loading model: {e}")
    model = None

def analyze_plant_image(image_bytes):
    """
    Takes raw bytes from the upload widget, runs AI,
    and returns (Disease Name, Confidence Score)
    """
    if model is None:
        return "AI Unavailable", 0.0

    try:
        # 1. Convert bytes to PIL Image
        image = Image.open(io.BytesIO(image_bytes)).convert("RGB")

        # 2. Preprocess
        inputs = processor(images=image, return_tensors="pt")

        # 3. Predict
        with torch.no_grad():
            outputs = model(**inputs)

        # 4. Get Result
        logits = outputs.logits
        predicted_class_idx = logits.argmax(-1).item()
        confidence = torch.nn.functional.softmax(logits, dim=-1)[0][predicted_class_idx].item()

        label = model.config.id2label[predicted_class_idx]

        return label, confidence
    except Exception as e:
        print(f"Prediction Error: {e}")
        return "Error analyzing", 0.0
````

````python
# @title 1. Global Setup & Database Functions
import google.generativeai as genai
import requests
import base64
import json
from PIL import Image
import io
from datetime import datetime

# ==========================================
# üîë CONFIGURATION
# ==========================================
GEMINI_API_KEY = "AIzaSyBEaBBufcG4W-2MN0Asufd-Gzx2iW5xku4"
FIREBASE_URL = "https://project-plant-diseases-default-rtdb.firebaseio.com"

# Configure Gemini
genai.configure(api_key=GEMINI_API_KEY)
plant_ai = genai.GenerativeModel("models/gemini-2.5-flash")

# ==========================================
# üíæ DATABASE HELPER FUNCTIONS
# ==========================================
def save_plant_to_db(plant_id, plant_data):
    """Saves a plant using simple HTTP PUT."""
    url = f"{FIREBASE_URL}/plants/{plant_id}.json"

    # Convert bytes to Base64 string for JSON storage if needed
    if isinstance(plant_data.get("image"), bytes):
        plant_data["image"] = base64.b64encode(plant_data["image"]).decode('utf-8')

    try:
        response = requests.put(url, json=plant_data)
        response.raise_for_status()
        return True
    except Exception as e:
        print(f"‚ùå Error saving to Firebase: {e}")
        return False

def get_all_plants_from_db():
    url = f"{FIREBASE_URL}/plants.json"
    try:
        response = requests.get(url)
        if response.status_code == 200 and response.json():
            return response.json()
        return {}
    except Exception as e:
        print(f"‚ùå Error fetching from Firebase: {e}")
        return {}

def delete_plant_from_db(plant_id):
    url = f"{FIREBASE_URL}/plants/{plant_id}.json"
    try:
        requests.delete(url)
    except Exception as e:
        print(f"‚ùå Error deleting from Firebase: {e}")

def update_plant_in_db(plant_id, updates):
    url = f"{FIREBASE_URL}/plants/{plant_id}.json"
    try:
        requests.patch(url, json=updates)
    except Exception as e:
        print(f"‚ùå Error updating Firebase: {e}")

def compress_image(image_bytes, max_size=(400, 400), quality=60):
    """
    Resizes image to max 400px and reduces JPEG quality.
    Returns: Compressed raw bytes.
    """
    try:
        # 1. Open image
        img = Image.open(io.BytesIO(image_bytes))

        # 2. Convert to RGB (in case it's PNG with transparency)
        if img.mode in ("RGBA", "P"):
            img = img.convert("RGB")

        # 3. Resize (maintain aspect ratio)
        img.thumbnail(max_size)

        # 4. Save to buffer with lower quality
        buffer = io.BytesIO()
        img.save(buffer, format="JPEG", quality=quality)

        return buffer.getvalue()
    except Exception as e:
        print(f"Compression failed: {e}")
        return image_bytes # Fallback to original if fail
````

````python
# @title Sensor Page

import requests
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from datetime import datetime, timedelta
from IPython.display import display, HTML
from collections import defaultdict

# ==================================================
# Styles
# ==================================================
SENSORS_HEADER_STYLE = widgets.HTML("""
<style>
.page { padding: 24px; }
.header {
    background: linear-gradient(135deg, #22c55e, #34d399);
    color: white; border-radius: 20px; padding: 32px; margin-bottom: 32px;
}
.mapreduce-card {
    background: white; border-radius: 16px; padding: 24px;
    margin-top: 24px; box-shadow: 0 8px 20px rgba(0,0,0,0.08);
}
/* ◊°◊í◊†◊ï◊ü ◊î◊ë◊ï◊¢◊ï◊™ (Chips) */
.chip-container {
    display: flex; flex-wrap: wrap; gap: 12px; margin-bottom: 20px;
}
.chip {
    display: flex; align-items: center; background: #f8fafc;
    border: 1px solid #e2e8f0; border-radius: 50px; padding: 8px 16px;
    transition: all 0.3s ease;
}
.chip:hover { transform: translateY(-2px); box-shadow: 0 4px 6px rgba(0,0,0,0.05); }
.chip-icon { font-size: 18px; margin-right: 10px; }
.chip-content { display: flex; flex-direction: column; }
.chip-label { font-size: 11px; color: #64748b; text-transform: uppercase; font-weight: bold; }
.chip-value { font-size: 14px; color: #1e293b; font-weight: 600; }
.chip-count {
    margin-left: 12px; background: #22c55e; color: white;
    border-radius: 12px; padding: 2px 10px; font-size: 12px; font-weight: bold;
}
</style>
"""
)

FIREBASE_URL = "https://project-plant-diseases-default-rtdb.firebaseio.com"
BASE_URL = "https://server-cloud-v645.onrender.com"

# ==================================================
# Firebase Saving Logic (FIXED & DEBUGGED)
# ==================================================
def save_snapshot_to_firebase(snapshot):
    today_date = datetime.now().strftime("%d-%m-%Y")
    entry_id = datetime.now().strftime("data%H%M%S")

    # Construct the payload
    # Note: We check if keys exist to avoid KeyErrors
    payload = {
        "temperature": snapshot["temperature"]["value"] if snapshot.get("temperature") else None,
        "humidity": snapshot["humidity"]["value"] if snapshot.get("humidity") else None,
        "soil_moisture": snapshot["soil"]["value"] if snapshot.get("soil") else None,
        "timestamp": datetime.now().isoformat()
    }

    url = f"{FIREBASE_URL}/sensor_data/{today_date}/{entry_id}.json"
    try:
        response = requests.put(url, json=payload)

        # üîç CHECK FOR ERRORS
        if response.status_code == 200:
            print(" ")
        else:
            print(f"‚ùå Firebase Error: {response.status_code}")
            print(f"‚ùå Response details: {response.text}")

    except Exception as e:
        print(f"‚ùå Network/Request Error: {e}")
# ==================================================
# üö® ALERT ENGINE (ACTIVE + HISTORY)
# ==================================================

THRESHOLDS = {
    "temperature": {"min": 24, "max": 25},
    "humidity": {"min": 28, "max": 35},
    "soil_moisture": {"min": 30, "max": 80}
}

def now_iso():
    return datetime.now().isoformat()

def get_active_alert(sensor):
    return requests.get(
        f"{FIREBASE_URL}/alerts/active/{sensor}.json"
    ).json()

def create_active_alert(sensor, value):
    payload = {
        "sensor": sensor,
        "value": value,
        "since": now_iso()
    }
    requests.put(
        f"{FIREBASE_URL}/alerts/active/{sensor}.json",
        json=payload
    )

def move_alert_to_history(sensor, alert):
    payload = {
        "sensor": sensor,
        "value": alert.get("value"),
        "start": alert.get("since"),
        "end": now_iso()
    }
    requests.post(
        f"{FIREBASE_URL}/alerts/history.json",
        json=payload
    )

def clear_active_alert(sensor):
    requests.delete(
        f"{FIREBASE_URL}/alerts/active/{sensor}.json"
    )

def process_alert(sensor, value):
    limits = THRESHOLDS[sensor]
    active_alert = get_active_alert(sensor)

    out_of_range = value < limits["min"] or value > limits["max"]

    # üî¥ ◊ó◊®◊ô◊í◊î ‚Üí ◊¶◊ï◊® ◊î◊™◊®◊ê◊î
    if out_of_range and not active_alert:
        create_active_alert(sensor, value)

    # üü¢ ◊ó◊ñ◊® ◊ú◊ò◊ï◊ï◊ó ‚Üí ◊î◊¢◊ë◊® ◊ú◊î◊ô◊°◊ò◊ï◊®◊ô◊î
    if not out_of_range and active_alert:
        move_alert_to_history(sensor, active_alert)
        clear_active_alert(sensor)


# ==================================================
# Fetch latest sensors (ROBUST VERSION)
# ==================================================
def get_latest_sensors():
    snapshot = {}
    for feed in ["temperature", "humidity", "soil"]:
        try:
            r = requests.get(f"{BASE_URL}/history",
                             params={"feed": feed, "limit": 1}, timeout=5)
            data = r.json()

            # Step 1: Get the raw item
            if data.get("data") and len(data["data"]) > 0:
                raw_item = data["data"][0]

                # Step 2: Force it to be a dictionary
                if isinstance(raw_item, dict):
                    # It's already good (e.g., {'value': 25, 'timestamp': ...})
                    snapshot[feed] = raw_item
                else:
                    # It's just a number/string (e.g., "25"), wrap it!
                    snapshot[feed] = {
                        "value": raw_item,
                        "timestamp": datetime.now().isoformat()
                    }
            else:
                snapshot[feed] = None

        except Exception as e:
            # If fetching fails, just set to None
            snapshot[feed] = None

    # Step 3: Save to Firebase (safely)
    if any(snapshot.values()):
        try:
            # We use a temporary safe snapshot for saving to avoid crashes
            safe_snapshot = {}
            for k, v in snapshot.items():
                if v and isinstance(v, dict):
                    safe_snapshot[k] = v
                else:
                     # Create a dummy dict if something is wrong, to prevent save crash
                    safe_snapshot[k] = {"value": 0}

            save_snapshot_to_firebase(safe_snapshot)

            # Run Alerts safely
            if snapshot.get("temperature"):
                process_alert("temperature", float(snapshot["temperature"]["value"]))
            if snapshot.get("humidity"):
                process_alert("humidity", float(snapshot["humidity"]["value"]))
            if snapshot.get("soil"):
                process_alert("soil_moisture", float(snapshot["soil"]["value"]))

        except Exception as e:
            print(f"‚ö†Ô∏è Warning: Background save/alert failed, but UI will load. Error: {e}")

    return snapshot


def create_sensors_header():
    return widgets.HTML("""
    <div class="page">
        <div class="header">
            <div style="font-size:14px;opacity:0.9;">AgriVision System</div>
            <div style="font-size:32px;font-weight:800;">Sensor Monitoring</div>
            <div style="font-size:14px;opacity:0.9;">
                Real-time & historical temperature, humidity and soil data
            </div>
        </div>
    </div>
    """
)

# ==================================================
# Date range helper
# ==================================================
def get_date_range(selection):
    today = datetime.now().date()
    if selection == "today":
        return [today]
    days = 7 if selection == "7d" else 30
    return [today - timedelta(days=i) for i in range(days)]

# ==================================================
# Graph functions
# ==================================================
import matplotlib.dates as mdates

def sensor_graph(feed_key, title, color, unit, date_range):
    plot_data = []
    for d in date_range:
        date_str = d.strftime("%d-%m-%Y")
        try:
            fb_data = requests.get(f"{FIREBASE_URL}/sensor_data/{date_str}.json", timeout=5).json()
            if not fb_data: continue
            for entry in fb_data.values():
                key_name = "soil_moisture" if feed_key == "soil" else feed_key
                if entry.get(key_name) is not None:
                    plot_data.append({
                        "time": pd.to_datetime(entry["timestamp"]),
                        "value": float(entry[key_name])
                    })
        except: continue

    if not plot_data:
        return

    df = pd.DataFrame(plot_data).sort_values("time")
    fig, ax = plt.subplots(figsize=(10, 4))

    # ◊¶◊ô◊ï◊® ◊î◊ß◊ï ◊¢◊ù ◊†◊ß◊ï◊ì◊ï◊™
    ax.plot(df["time"], df["value"], color=color, linewidth=2.5,
            marker='o', markersize=5, markerfacecolor='white', markeredgewidth=1.5,
            label=title)

    ax.fill_between(df["time"], df["value"], color=color, alpha=0.1)

    # ◊î◊í◊ì◊®◊™ ◊¶◊ô◊® ◊î◊ñ◊û◊ü - ◊ô◊™◊ó◊ô◊ú ◊ê◊ï◊ò◊ï◊û◊ò◊ô◊™ ◊û◊î◊û◊ì◊ô◊ì◊î ◊î◊®◊ê◊©◊ï◊†◊î
    # ◊ß◊ï◊ë◊¢ ◊ê◊ï◊ò◊ï◊û◊ò◊ô◊™ ◊ú◊§◊ô ◊î◊†◊™◊ï◊†◊ô◊ù ◊¢◊¶◊û◊ù
    all_dates = df["time"].dt.date.unique()

    if len(all_dates) == 1:
        # ◊õ◊ú ◊î◊†◊™◊ï◊†◊ô◊ù ◊û◊ê◊ï◊™◊ï ◊ô◊ï◊ù ‚Üí ◊©◊¢◊ï◊™
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        plt.margins(x=0.02)
    else:
        # ◊†◊™◊ï◊†◊ô◊ù ◊û◊õ◊û◊î ◊ô◊û◊ô◊ù ‚Üí ◊™◊ê◊®◊ô◊õ◊ô◊ù
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))


    ax.set_title(f" Monitoring {title}", fontsize=14, fontweight='bold', pad=15)
    ax.set_ylabel(unit, fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

def combined_graph(date_range):
    fig, ax = plt.subplots(figsize=(12, 5))
    has_data = False

    configs = [
        ("temperature", "Temperature (¬∞C)", "#e67e22"),
        ("humidity", "Humidity (%)", "#3498db"),
        ("soil_moisture", "Soil Moisture (%)", "#27ae60")
    ]

    for key, label, color in configs:
        plot_data = []

        for d in date_range:
            try:
                fb_data = requests.get(
                    f"{FIREBASE_URL}/sensor_data/{d.strftime('%d-%m-%Y')}.json",
                    timeout=5
                ).json()

                if not fb_data:
                    continue

                for entry in fb_data.values():
                    if entry.get(key) is not None:
                        plot_data.append({
                            "time": pd.to_datetime(entry["timestamp"]),
                            "value": float(entry[key])
                        })
            except:
                continue

        if plot_data:
            has_data = True
            df = pd.DataFrame(plot_data).sort_values("time")

            ax.plot(
                df["time"],
                df["value"],
                label=label,
                color=color,
                linewidth=3.5,
                linestyle='-',
                marker='o',
                markersize=5,
                markerfacecolor='white',
                markeredgewidth=1.8
            )

    if not has_data:
        return

    # ‚è±Ô∏è ◊¶◊ô◊® ◊ñ◊û◊ü ◊ë◊®◊ï◊®
    if len(date_range) == 1:
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    else:
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))

    ax.set_title(
        "Combined Environmental Overview",
        fontsize=16,
        fontweight='bold',
        pad=20
    )

    ax.set_ylabel("Sensor Values", fontsize=12)

    # ◊í◊®◊ô◊ì ◊¢◊ì◊ô◊ü
    ax.grid(True, linestyle='--', alpha=0.35)

    # Legend ◊ë◊®◊ï◊® ◊ë◊§◊†◊ô◊ù
    ax.legend(
        loc='upper center',
        bbox_to_anchor=(0.5, -0.15),
        ncol=3,
        frameon=False,
        fontsize=11
    )

    all_dates = df["time"].dt.date.unique()

    if len(all_dates) == 1:
        start_time = df["time"].min()   # ◊î◊ß◊®◊ô◊ê◊î ◊î◊®◊ê◊©◊ï◊†◊î ◊ë-DB
        now = pd.Timestamp.now()        # ◊¢◊õ◊©◊ô◊ï

        ax.set_xlim(start_time, now)


    plt.tight_layout()
    plt.show()

# ==================================================
# üîµ MapReduce functions (ADDED)
# ==================================================
def map_temperature(v):
    v = float(v)
    # ◊ò◊ï◊ï◊ó: ◊û◊™◊ó◊™ ◊ú-18 ◊ñ◊î ◊†◊û◊ï◊ö, ◊ë◊ô◊ü 18 ◊ú-30 ◊ñ◊î ◊™◊ß◊ô◊ü, ◊û◊¢◊ú 30 ◊ñ◊î ◊í◊ë◊ï◊î
    return "Low" if v < 18 else "Normal" if v <= 30 else "High"

def map_humidity(v):
   v = float(v)
   # ◊ë◊ú◊ó◊ï◊™ ◊ê◊ï◊ï◊ô◊®: ◊û◊™◊ó◊™ ◊ú-30 ◊ñ◊î ◊†◊û◊ï◊ö, ◊¢◊ì 70 ◊ñ◊î ◊™◊ß◊ô◊ü
   return "Low" if v < 30 else "Normal" if v <= 70 else "High"

def map_soil(v):
    v = float(v)
    # ◊ë◊ú◊ó◊ï◊™ ◊ß◊®◊ß◊¢: ◊û◊™◊ó◊™ ◊ú-20 ◊ñ◊î ◊ô◊ë◊© (Dry), ◊¢◊ì 40 ◊ñ◊î ◊ê◊ï◊§◊ò◊ô◊û◊ú◊ô
    return "Dry" if v < 20 else "Optimal" if v <= 40 else "Wet"

def run_mapreduce(records, field, map_func, categories):
   counts = defaultdict(int)
   for r in records:
       if r.get(field) is not None:
          counts[map_func(r[field])] += 1
   return [(c, counts.get(c, 0)) for c in categories]


def fetch_records(date_range):
    records = []
    for d in date_range:
        fb_data = requests.get(
            f"{FIREBASE_URL}/sensor_data/{d.strftime('%d-%m-%Y')}.json"
        ).json()
        if fb_data:
            records.extend(fb_data.values())
    return records

# ==================================================
#◊§◊ï◊†◊ß◊¶◊ô◊ô◊™ ◊¢◊ñ◊® ◊ú◊ô◊ô◊¶◊ï◊® ◊î BAR CHART
# ==================================================

def create_sensor_bar_charts(res_t, res_h, res_s):
    data_sets = [res_t, res_h, res_s]
    titles = ["Temperature", "Humidity", "Soil Moisture"]

    color_map = {
        "Low": "#f1c40f",
        "Normal": "#2ecc71",
        "High": "#e74c3c",
        "Optimal": "#2ecc71",
        "Dry": "#f1c40f",
        "Wet": "#e74c3c"
    }

    fig, axs = plt.subplots(1, 3, figsize=(15, 5))

    for i, data in enumerate(data_sets):
        labels = [item[0] for item in data]
        values = [item[1] for item in data]

        current_colors = [color_map.get(label, "#b2bec3") for label in labels]

        bars = axs[i].bar(
            labels,
            values,
            color=current_colors,
            edgecolor='#2d3436',
            linewidth=1.2
        )

        # ◊¢◊®◊ö ◊û◊¢◊ú ◊õ◊ú ◊¢◊û◊ï◊ì◊î
        for bar in bars:
            height = bar.get_height()
            axs[i].text(
                bar.get_x() + bar.get_width() / 2,
                height + 0.2,
                f"{int(height)}",
                ha='center',
                va='bottom',
                fontsize=11,
                fontweight='bold'
            )

        axs[i].set_title(titles[i], fontweight='bold', fontsize=14, pad=15)

        # ◊¶◊ô◊® Y ◊ë◊®◊ï◊®
        axs[i].set_ylim(0, max(values) * 1.25 if values else 5)
        axs[i].set_ylabel("Count")

        # Grid ◊®◊ß ◊ë◊¶◊ô◊® Y
        axs[i].grid(axis='y', linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()
#=====================================
#for the temp , hum, soil
#=====================================

def get_latest_from_firebase():
    today = datetime.now().strftime("%d-%m-%Y")
    try:
        data = requests.get(
            f"{FIREBASE_URL}/sensor_data/{today}.json",
            timeout=5
        ).json()

        if not data:
            return {}

        # ◊ß◊ó ◊ê◊™ ◊î◊®◊©◊ï◊û◊î ◊î◊ê◊ó◊®◊ï◊†◊î ◊ú◊§◊ô timestamp
        latest = max(
            data.values(),
            key=lambda x: x.get("timestamp", "")
        )

        return {
            "temperature": latest.get("temperature", "--"),
            "humidity": latest.get("humidity", "--"),
            "soil": latest.get("soil_moisture", "--")
        }

    except Exception as e:
        print("‚ö†Ô∏è Firebase fetch failed:", e)
        return {}

# ==================================================
# UI Component ‚Äì Sensor Card
# ==================================================
def sensor_card(icon, title, value, unit, color):
    return widgets.HTML(f"""
    <div style="
        background:white;
        border-radius:16px;
        padding:18px;
        width:220px;
        box-shadow:0 8px 20px rgba(0,0,0,0.08);
        text-align:center;
    ">
        <div style="font-size:32px">{icon}</div>
        <div style="color:#7f8c8d;font-size:14px;margin-top:6px">
            {title}
        </div>
        <div style="font-size:26px;font-weight:700;color:{color}">
            {value} <span style="font-size:16px">{unit}</span>
        </div>
    </div>
    """)

# ==================================================
# Render Sensor Page (FIXED)
# ==================================================
def render_sensors():
    # 1. Fetch live sensors (This triggers the Save and Alerts)
    live_sensors = get_latest_sensors()

    # 2. Optimization: Don't fetch from Firebase immediately.
    # Use the live data we just got for the cards.
    # We only format it to match the 'value' structure if needed.

    # Simple helper to extract value for the card
    # 2. Smart Helper: Handles both Dicts and Raw Values
    def get_val(key):
        item = live_sensors.get(key)

        # Case A: Data is missing
        if item is None:
            return "--"

        # Case B: It's a dictionary (e.g., {'value': 25, 'timestamp': ...})
        if isinstance(item, dict):
            return item.get("value", "--")

        # Case C: It's just a raw number/string (e.g., 25)
        return item

    # 1. State management
    state = {'cached_records': None, 'current_range': None}

    # Use the LIVE data for the cards, not the historical fetch
    cards_row = widgets.HBox([
      sensor_card("üå°Ô∏è", "Temperature", get_val("temperature"), "¬∞C", "#e67e22"),
      sensor_card("üíß", "Humidity", get_val("humidity"), "%", "#3498db"),
      sensor_card("üå±", "Soil", get_val("soil"), "%", "#27ae60")
    ], layout=widgets.Layout(justify_content="center", gap="20px"))

    range_selector = widgets.Dropdown(
        options=[("Today","today"),("Last 7 Days","7d"),("Last 30 Days","30d")],
        value="today",
        description="üìÖ Range:"
    )

    chk_temp = widgets.Checkbox(description="üå°Ô∏è Temperature", value=False)
    chk_hum  = widgets.Checkbox(description="üíß Humidity", value=False)
    chk_soil = widgets.Checkbox(description="üå± Soil Moisture", value=False)

    graph_container = widgets.Output()
    mapreduce_container = widgets.Output()

    def update_view(change=None):
        date_range = get_date_range(range_selector.value)

        # 2. Centralized Data Fetch for GRAPHS (History)
        if state['current_range'] != range_selector.value:
            with mapreduce_container:
                mapreduce_container.clear_output()
                display(HTML("<p style='text-align:center; color:#27ae60;'>üîÑ Syncing data from Firebase...</p>"))

            state['cached_records'] = fetch_records(date_range)
            state['current_range'] = range_selector.value

        records = state['cached_records']

        # 3. Update Line Graphs
        with graph_container:
            graph_container.clear_output()
            if not (chk_temp.value or chk_hum.value or chk_soil.value):
                combined_graph(date_range)
            else:
                if chk_temp.value: sensor_graph("temperature","Temperature","#e67e22","¬∞C",date_range)
                if chk_hum.value:  sensor_graph("humidity","Humidity","#3498db","%",date_range)
                if chk_soil.value: sensor_graph("soil","Soil","#27ae60","%",date_range)

        # 4. Update Bar Charts
        with mapreduce_container:
            mapreduce_container.clear_output()
            if records:
                res_t = run_mapreduce(records, "temperature", map_temperature, ["Low", "Normal", "High"])
                res_h = run_mapreduce(records, "humidity", map_humidity, ["Low", "Normal", "High"])
                res_s = run_mapreduce(records, "soil_moisture", map_soil, ["Dry", "Optimal", "Wet"])

                display(HTML("<h3 style='text-align:center; color:#2d3436;'>üìä (MapReduce)</h3>"))
                create_sensor_bar_charts(res_t, res_h, res_s)
            else:
                display(HTML("<p style='text-align:center; color:#e74c3c;'>‚ö†Ô∏è No data found for this range.</p>"))

    # Observe all widgets
    for w in [range_selector, chk_temp, chk_hum, chk_soil]:
        w.observe(update_view, names="value")

    update_view()

    return widgets.VBox([
        SENSORS_HEADER_STYLE,
        create_sensors_header(),
        cards_row,
        range_selector,
        widgets.HBox([chk_temp, chk_hum, chk_soil]),
        graph_container,
        mapreduce_container
    ])
````

````python
# @title Rewards Page (Microservices Architecture)
import ipywidgets as widgets
from IPython.display import display, clear_output
import requests
import json
import uuid
from datetime import datetime
import google.generativeai as genai

# =====================================================
# 1. CONFIGURATION & SHARED UTILS
# =====================================================
FIREBASE_URL = "https://project-plant-diseases-default-rtdb.firebaseio.com"
GEMINI_API_KEY = "AIzaSyBEaBBufcG4W-2MN0Asufd-Gzx2iW5xku4"

genai.configure(api_key=GEMINI_API_KEY)

# Shared Styles
STYLE = widgets.HTML("""
<style>
    .task-btn:hover { transform: scale(1.02); opacity: 0.9; cursor: pointer; }
    .service-badge { font-size: 10px; padding: 2px 6px; border-radius: 4px; text-transform: uppercase; font-weight: bold; margin-right: 8px; }
    .badge-iot { background: #e0f2fe; color: #0369a1; border: 1px solid #bae6fd; }
    .badge-botanist { background: #f0fdf4; color: #15803d; border: 1px solid #bbf7d0; }
</style>
""")

# =====================================================
# 2. DATA ACCESS LAYER (DAL)
# =====================================================

def get_latest_sensors2():
    """Fetches sensor data and ensures we only get the VALUES, not the whole JSON object."""
    try:
        today = datetime.now().strftime("%d-%m-%Y")
        url = f"{FIREBASE_URL}/sensor_data/{today}.json"
        data = requests.get(url, timeout=3).json()

        # Default fallback
        defaults = {"temperature": 25, "humidity": 60, "soil": 500}

        if not data:
            return defaults

        # Get most recent reading entry
        latest = sorted(data.values(), key=lambda x: x.get("timestamp", ""), reverse=True)[0]

        # Helper to extract value if it's nested in a dict
        def extract_val(val, default):
            if isinstance(val, dict):
                return val.get('value', default)
            return val if val is not None else default

        return {
            "temperature": extract_val(latest.get("temperature"), 25),
            "humidity": extract_val(latest.get("humidity"), 60),
            "soil": extract_val(latest.get("soil_moisture"), 500) # Note: check if key is 'soil' or 'soil_moisture' in your DB
        }
    except Exception as e:
        print(f"Sensor Fetch Error: {e}")
        return {"temperature": 25, "humidity": 60, "soil": 500}

def get_saved_plants():
    """Fetches all plants saved in 'My Garden'."""
    try:
        response = requests.get(f"{FIREBASE_URL}/plants.json").json()
        return response if response else {}
    except:
        return {}

def get_user_gamification():
    """Fetches user XP and Tasks."""
    try:
        response = requests.get(f"{FIREBASE_URL}/gamification.json").json()
        if not response:
            init_data = {"level": 1, "current_xp": 0, "next_level_xp": 500, "last_active_date": "", "tasks": []}
            requests.put(f"{FIREBASE_URL}/gamification.json", json=init_data)
            return init_data
        return response
    except:
        return None

def update_xp(points):
    """Adds XP and handles leveling up."""
    data = get_user_gamification()
    if not data: return

    data['current_xp'] += points
    if data['current_xp'] >= data['next_level_xp']:
        data['level'] += 1
        data['current_xp'] -= data['next_level_xp']
        data['next_level_xp'] = int(data['next_level_xp'] * 1.2)

    requests.patch(f"{FIREBASE_URL}/gamification.json", json={
        "level": data['level'],
        "current_xp": data['current_xp'],
        "next_level_xp": data['next_level_xp']
    })
    return data

# =====================================================
# 3. MICROSERVICES (The Brains)
# =====================================================

def service_iot_tasks(sensor_data):
    """
    SERVICE 1: IoT MONITOR
    Generates 4 tasks based strictly on environment data.
    """
    model = genai.GenerativeModel('gemini-2.5-flash')

    prompt = f"""
    Role: IoT Environmental Controller.
    Data: Temp={sensor_data['temperature']}C, Humidity={sensor_data['humidity']}%, Soil={sensor_data['soil']}.

    Task: Generate exactly 4 short tasks to optimize this environment.
    Focus: Watering, Ventilation, Heating, Cooling.

    Output JSON ONLY: [{{ "task": "string", "points": int (1-5) }}]
    """

    try:
        res = model.generate_content(prompt)
        tasks = json.loads(res.text.replace('```json', '').replace('```', '').strip())
        # Tag the tasks with source
        for t in tasks: t['source'] = 'IOT'
        return tasks[:4] # Ensure strict limit
    except Exception as e:
        print(f"IoT Service Error: {e}")
        return []

def service_garden_tasks(plants_dict):
    """
    SERVICE 2: BOTANIST
    Generates 4 tasks based on specific saved plants.
    """
    if not plants_dict:
        return [{"task": "Scan your first plant in the Garden tab!", "points": 5, "source": "GARDEN"}]

    # Summarize plant list for token efficiency
    plant_summary = []
    for pid, p in list(plants_dict.items())[:5]: # Limit to 5 plants for context window
        name = p.get('name', 'Unknown Plant')
        health = p.get('health', 100)
        diagnosis = p.get('diagnosis', 'Healthy')
        plant_summary.append(f"- {name} (Health: {health}%, Issue: {diagnosis})")

    summary_str = "\n".join(plant_summary)

    model = genai.GenerativeModel('gemini-2.5-flash')

    prompt = f"""
    Role: Expert Botanist.
    My Garden Plants:
    {summary_str}

    Task: Generate exactly 4 short care tasks specifically for these plants.
    Focus: Pruning, specific treatments, inspecting leaves, fertilizing specific species.

    Output JSON ONLY: [{{ "task": "string", "points": int (1-5) }}]
    """

    try:
        res = model.generate_content(prompt)
        tasks = json.loads(res.text.replace('```json', '').replace('```', '').strip())
        for t in tasks: t['source'] = 'GARDEN'
        return tasks[:4]
    except Exception as e:
        print(f"Garden Service Error: {e}")
        return []

# =====================================================
# 4. ORCHESTRATOR (Task Manager)
# =====================================================

def generate_daily_manifest(existing_tasks):
    """
    Calls both services and combines tasks if it's a new day.
    """
    sensors = get_latest_sensors2()
    plants = get_saved_plants()
    today_str = datetime.now().strftime("%Y-%m-%d")

    # Filter: Keep incomplete tasks from yesterday
    active_tasks = [t for t in existing_tasks if not t['done']]

    print("üîÑ Contacting IoT Service...")
    iot_tasks = service_iot_tasks(sensors)

    print("üîÑ Contacting Garden Service...")
    garden_tasks = service_garden_tasks(plants)

    new_batch = iot_tasks + garden_tasks

    # Process new tasks with IDs
    for t in new_batch:
        active_tasks.append({
            "id": str(uuid.uuid4()),
            "task": t['task'],
            "points": t.get('points', 1),
            "source": t.get('source', 'GENERIC'), # IOT or GARDEN
            "done": False,
            "date_created": today_str
        })

    # Sort: High points first
    active_tasks.sort(key=lambda x: x['points'], reverse=True)

    # Save
    requests.patch(f"{FIREBASE_URL}/gamification.json", json={
        "tasks": active_tasks,
        "last_active_date": today_str
    })

    return active_tasks

# =====================================================
# 5. UI COMPONENTS (UPDATED)
# =====================================================

def render_hero(user_data):
    percent = int((user_data["current_xp"] / user_data["next_level_xp"]) * 100)
    return widgets.HTML(f"""
    <div style="
        background: linear-gradient(135deg,#22c55e,#4ade80);
        border-radius:24px;
        padding:30px;
        color:white;
        box-shadow:0 10px 30px rgba(34,197,94,0.35);
    ">
        <div style="display:flex; justify-content:space-between; align-items:center;">
            <div>
                <div style="text-transform:uppercase; font-size:12px; letter-spacing:1px; opacity:0.85;">
                    Gardener Level
                </div>
                <div style="font-size:42px; font-weight:800;">
                    {user_data['level']}
                </div>
            </div>
            <div style="text-align:right;">
                <div style="font-size:24px; font-weight:bold;">
                    {user_data['current_xp']}
                    <span style="font-size:14px; opacity:0.85;">XP</span>
                </div>
                <div style="font-size:12px; opacity:0.85;">
                    Next Level: {user_data['next_level_xp']} XP
                </div>
            </div>
        </div>

        <div style="
            margin-top:20px;
            background:rgba(255,255,255,0.25);
            height:8px;
            border-radius:10px;
            overflow:hidden;
        ">
            <div style="
                width:{percent}%;
                background:white;
                height:100%;
            "></div>
        </div>
    </div>
    """)

def render_task_list(tasks, refresh_callback):
    items = []

    # Header with Debug Button
    header_box = widgets.HBox([
        widgets.HTML("<h3 style='margin:0; color:#333;'>Today's Missions</h3>"),
        # Button to FORCE regeneration
    ], layout=widgets.Layout(justify_content='space-between', align_items='center', margin='20px 0 10px 0'))

    items.append(header_box)

    if not tasks:
        items.append(widgets.HTML("<div style='color:#777;'>All tasks complete! Great job!</div>"))

    for task in tasks:
        is_done = task.get('done', False)
        source = task.get('source', 'GENERIC')

        # Badge Style
        badge_class = "badge-iot" if source == 'IOT' else "badge-botanist" if source == 'GARDEN' else ""
        badge_html = f"<span class='service-badge {badge_class}'>{source}</span>" if badge_class else ""

        opacity = "0.5" if is_done else "1.0"
        bg = "#f3f4f6" if is_done else "white"

        task_html = widgets.HTML(f"""
        <div style="display:flex; align-items:center; justify-content:space-between; padding:15px;
                    background:{bg}; border-radius:12px; margin-bottom:10px; border:1px solid #eee; opacity:{opacity};">
            <div style="display:flex; align-items:center; gap:15px;">
                <div style="background:{'#22c55e' if not is_done else '#9ca3af'}; color:white; width:30px; height:30px; border-radius:50%;
                            display:flex; align-items:center; justify-content:center; font-weight:bold; font-size:12px;">
                    {task['points']}
                </div>
                <div>
                    <div>{badge_html} <span style="font-weight:600; color:#1f2937;">{task['task']}</span></div>
                </div>
            </div>
        </div>
        """)

        btn = widgets.Button(
            description="Claim" if not is_done else "Done",
            disabled=is_done,
            button_style='success' if is_done else '',
            layout=widgets.Layout(width='80px')
        )
        if not is_done:
            btn.style.button_color = '#4f46e5'
            btn.style.text_color = 'white'

        def on_click(b, t_id=task['id'], pts=task['points']):
            user_data = get_user_gamification()
            for t in user_data['tasks']:
                if t.get('id') == t_id:
                    t['done'] = True
                    break
            requests.patch(f"{FIREBASE_URL}/gamification.json", json={"tasks": user_data['tasks']})
            update_xp(pts)
            refresh_callback()

        btn.on_click(on_click)
        row = widgets.HBox([task_html, btn], layout=widgets.Layout(align_items='center', justify_content='space-between', width='100%'))
        items.append(row)

    return widgets.VBox(items)

# =====================================================
# 6. MAIN APP (UPDATED WITH FORCE REFRESH)
# =====================================================

def main_app_microservices():
    out = widgets.Output()

    def run_generation():
        """Forces the generation of new tasks."""
        with out:
            out.clear_output(wait=True)
            print("üîÑ Microservices Running: Analyzing IoT sensors & Garden DB...")

        # 1. Fetch current data
        user_data = get_user_gamification()
        existing_tasks = user_data.get('tasks', [])

        # 2. Run Logic
        new_tasks = generate_daily_manifest(existing_tasks)

        # 3. Refresh UI
        refresh_ui()

    def refresh_ui():
        with out:
            out.clear_output(wait=True)
            fresh_data = get_user_gamification()

            display(STYLE)
            display(render_hero(fresh_data))

            # Regeneration Button
            regen_btn = widgets.Button(description="‚Üª Regenerate Tasks", icon="refresh")
            regen_btn.on_click(lambda b: run_generation())
            display(widgets.HBox([widgets.HTML("<b>Debug Tool:</b> "), regen_btn], layout=widgets.Layout(margin='0 0 20px 0')))

            s = get_latest_sensors2()
            display(widgets.HTML(f"""
            <div style="margin-top:10px; font-size:13px; color:#666; text-align:center;">
                üì° <b>Live IoT Data:</b> {s['temperature']}¬∞C | {s['humidity']}% Humidity | Soil: {s['soil']}
            </div>
            """))

            display(render_task_list(fresh_data['tasks'], refresh_ui))

    # Initial Run
    refresh_ui()
    return out
````

````python
# @title Alerts Page ‚Äì FINAL (Plant Alerts Only)

import requests
import ipywidgets as widgets
from datetime import datetime
from IPython.display import display

# =====================================================
# üîë CONFIG
# =====================================================
FIREBASE_URL = "https://project-plant-diseases-default-rtdb.firebaseio.com"

# =====================================================
# üì• FETCH PLANTS
# =====================================================
def fetch_plants():
    try:
        return requests.get(f"{FIREBASE_URL}/plants.json").json() or {}
    except:
        return {}

# =====================================================
# üß† ANALYZE PLANT STATUS
# =====================================================
def analyze_plant_status(plant):
    health = plant.get("health", 100)

    if health >= 85:
        return "safe", "üå± Plant is healthy ‚Äì keep current care"
    elif health >= 70:
        return "safe", "üå± Plant is doing well ‚Äì continue monitoring"
    elif health >= 50:
        return "warning", "‚ö† Plant health is declining ‚Äì needs attention"
    elif health >= 30:
        return "critical", "‚ùó Plant needs care ‚Äì action recommended"
    else:
        return "critical", "‚ùó Plant in critical condition ‚Äì urgent action required"

# =====================================================
# üíæ SAVE / UPDATE PLANT ALERT (NO DUPLICATES)
# =====================================================
def save_plant_alert(plant_id, plant):
    status, message = analyze_plant_status(plant)

    payload = {
        "plant_id": plant_id,
        "plant_name": plant.get("name", "Plant"),
        "health": plant.get("health", 0),
        "status": status,
        "message": message,
        "created_at": datetime.now().isoformat(),
        "source": "plant_health"
    }

    # plant_id ◊õ÷æKEY ‚Üí ◊ê◊ô◊ü ◊õ◊§◊ô◊ú◊ï◊ô◊ï◊™
    requests.put(
        f"{FIREBASE_URL}/alerts/active/{plant_id}.json",
        json=payload
    )

# =====================================================
# üö® GENERATE ALERTS FROM PLANTS
# =====================================================
def generate_alerts_from_plants():
    plants = fetch_plants()
    for plant_id, plant in plants.items():
        save_plant_alert(plant_id, plant)

# =====================================================
# üì• FETCH ALERTS (PLANTS ONLY)
# =====================================================
def fetch_alerts():
    try:
        data = requests.get(f"{FIREBASE_URL}/alerts/active.json").json() or {}
        alerts = []

        for alert in data.values():
            # ◊û◊°◊†◊ü ◊õ◊ú ◊ì◊ë◊® ◊©◊ê◊ô◊†◊ï Plant Alert
            if alert.get("source") == "plant_health" and "status" in alert:
                alerts.append(alert)

        alerts.sort(
            key=lambda x: x.get("created_at", ""),
            reverse=True
        )

        return alerts
    except:
        return []

# =====================================================
# üé® STYLES
# =====================================================
STYLE = widgets.HTML("""
<style>
.page {
    max-width: 1000px;
    margin: auto;
    font-family: Inter, Arial;
}

.header {
    background: linear-gradient(135deg, #22c55e, #4ade80);
    color: white;
    border-radius: 22px;
    padding: 32px;
    margin-bottom: 32px;
}

.header-title {
    font-size: 28px;
    font-weight: 900;
}

.header-sub {
    font-size: 14px;
    opacity: 0.95;
}

.alert {
    border-radius: 18px;
    padding: 18px 22px;
    margin-bottom: 18px;
}

.alert.safe {
    background: #ecfdf5;
    border-left: 6px solid #22c55e;
    color: #065f46;
}

.alert.warning {
    background: #fffbeb;
    border-left: 6px solid #facc15;
    color: #92400e;
}

.alert.critical {
    background: #fef2f2;
    border-left: 6px solid #ef4444;
    color: #7f1d1d;
}

.alert-title {
    font-weight: 900;
    font-size: 16px;
}

.alert-body {
    font-size: 14px;
    margin-top: 6px;
}
</style>
""")

# =====================================================
# üñ•Ô∏è RENDER ALERTS PAGE
# =====================================================
def render_alerts():
    generate_alerts_from_plants()
    alerts = fetch_alerts()
    cards = []

    for alert in alerts:
        cards.append(widgets.HTML(f"""
        <div class="alert {alert['status']}">
            <div class="alert-title">üå± {alert['plant_name']}</div>
            <div class="alert-body">
                {alert['message']}<br>
                Health score: <b>{alert['health']}%</b>
            </div>
        </div>
        """))

    return widgets.VBox([
        STYLE,
        widgets.HTML('<div class="page">'),
        widgets.HTML("""
        <div class="header">
            <div class="header-title">üåø Plant Alerts</div>
            <div class="header-sub">Real-time health status for all your plants</div>
        </div>
        """),
        widgets.VBox(cards),
        widgets.HTML('</div>')
    ])
````
